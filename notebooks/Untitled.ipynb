```python
# ============================================
# NOTEBOOK EXPLORATORIO - SISTEMA HOSPITALARIO
# ============================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Carga de los 4 datasets originales
consultas = pd.read_parquet("C:/Users/pc/evaluacion-kedro/evaluacion/data/02_intermediate/consultas_cleaned.parquet")
examenes = pd.read_parquet("C:/Users/pc/evaluacion-kedro/evaluacion/data/02_intermediate/examenes_cleaned.parquet")
medicamentos = pd.read_parquet("C:/Users/pc/evaluacion-kedro/evaluacion/data/02_intermediate/medicamentos_cleaned.parquet")
pacientes = pd.read_parquet("C:/Users/pc/evaluacion-kedro/evaluacion/data/02_intermediate/pacientes_cleaned.parquet")

print("✅ Datasets cargados correctamente!")
print(f"consultas: {consultas.shape}")
print(f"examenes: {examenes.shape}")
print(f"medicamentos: {medicamentos.shape}")
print(f"pacientes: {pacientes.shape}")
```

```python
# ============================================
# PASO 2: EXPLORACIÓN INICIAL
# ============================================
for nombre, df in [("consultas", consultas), ("examenes", examenes),
                   ("medicamentos", medicamentos), ("pacientes", pacientes)]:
    print(f"\n{'='*40}")
    print(f"Dataset: {nombre}")
    print(f"Filas: {df.shape[0]} | Columnas: {df.shape[1]}")
    print(f"Nulos restantes: {df.isnull().sum().sum()}")
    print(df.dtypes)
```

```python
# ============================================
# PASO 3: ESTADÍSTICAS DESCRIPTIVAS
# ============================================
print("Estadísticas de consultas:")
print(consultas.describe())
```

```python
# ============================================
# PASO 4: VISUALIZACIÓN FINAL
# ============================================
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Análisis Exploratorio - Sistema Hospitalario", fontsize=16, fontweight='bold')

# Gráfico 1: Consultas por especialidad
especialidades = consultas['especialidad'].value_counts().head(8)
axes[0, 0].barh(especialidades.index, especialidades.values, color='#4ECDC4')
axes[0, 0].set_title("Top 8 Especialidades")
axes[0, 0].set_xlabel("Cantidad")

# Gráfico 2: Distribución de costos
consultas['costo'].dropna().astype(float).plot(kind='hist', bins=20, ax=axes[0, 1], color='#FF6B6B')
axes[0, 1].set_title("Distribución de Costos")
axes[0, 1].set_xlabel("Costo")

# Gráfico 3: Tipos de examen
tipos = examenes['tipo_examen'].value_counts().head(6)
axes[0, 2].bar(tipos.index, tipos.values, color='#45B7D1')
axes[0, 2].set_title("Tipos de Examen")
axes[0, 2].tick_params(axis='x', rotation=45)

# Gráfico 4: Distribución por género
genero = pacientes['genero'].value_counts()
axes[1, 0].pie(genero.values, labels=genero.index, autopct='%1.1f%%', colors=['#FF6B6B', '#4ECDC4'])
axes[1, 0].set_title("Distribución por Género")

# Gráfico 5: Top medicamentos
meds = medicamentos['medicamento'].value_counts().head(8)
axes[1, 1].barh(meds.index, meds.values, color='#96CEB4')
axes[1, 1].set_title("Top 8 Medicamentos")

# Gráfico 6: Pacientes por previsión
prevision = pacientes['prevision'].value_counts()
axes[1, 2].bar(prevision.index, prevision.values, color='#FFEAA7')
axes[1, 2].set_title("Pacientes por Previsión")
axes[1, 2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig("C:/Users/pc/evaluacion-kedro/evaluacion/data/08_reporting/visualizacion_final.png", dpi=150)
plt.show()
print("✅ Visualización guardada!")
```
